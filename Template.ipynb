## 0.2 Equations

You can take advantage of LaTeX when writing equations. If you've never used LaTeX before, the most important thing to know is how to pronounce it: you say "lay-teck" (or some people say "lah-teck") but not "lay-tecks" like the material. You write a formula using a \$ sign to begin and end the equation.

Here is an example:
$ f(x) = \frac{x^2}{2} $

This is entered as
```
$ f(x) = \frac{x^2}{2} $
```

Here are some other more complicated examples (double click in this cell to see how they are typeset):
